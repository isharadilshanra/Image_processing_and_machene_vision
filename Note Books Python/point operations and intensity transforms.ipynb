```python
# Intensity Windowing

%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

c = np.array([(100, 50), (150, 200)])

t1 = np.linspace(0, c[0,1], c[0,0] + 1 - 0).astype('uint8')
print(len(t1))
t2 = np.linspace(c[0,1] + 1, c[1,1], c[1,0] - c[0,0]).astype('uint8')
print(len(t2))
t3 = np.linspace(c[1,1] + 1, 255, 255 - c[1,0]).astype('uint8')
print(len(t3)) 
 
transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
print(len(transform))

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal') 
# plt.savefig('../../EN2550Lectures/en2550_lec02_point_operations_and_linear_filtering/figures/intensity_windowing.png')
plt.show()

img_orig = cv.imread('prop 1.jpg', cv.IMREAD_GRAYSCALE)
cv.namedWindow("Image")
cv.imshow("Image", img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig, transform)
cv.imshow("Image", image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Histogram of a Grayscale Image
%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('prop 1.jpg', cv.IMREAD_GRAYSCALE)
hist = cv.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

cv.calcHist(): This function is part of the OpenCV library and is used to calculate the histogram of an image.

[img]: This is a list containing the input image img for which you want to calculate the histogram.

[0]: This is a list specifying the channel index for which you want to calculate the histogram. In this case, [0] indicates that you are calculating the histogram for the first channel of the image. For color images in OpenCV, the channels are usually indexed as 0 for blue, 1 for green, and 2 for red.

None: This parameter specifies the mask. A mask can be used to calculate the histogram for specific regions of the image. When set to None, the histogram is calculated for the entire image.

[256]: This parameter specifies the number of bins in the histogram. Bins are intervals that divide the intensity range (0 to 255 in the case of an 8-bit image) into segments. In this case, you're using 256 bins, which means each possible intensity value corresponds to its own bin.

[0, 256]: This parameter specifies the range of values for which the histogram will be calculated. In this case, the range is from 0 to 255, which is the entire intensity range of an 8-bit image.

The result of the cv.calcHist() function is stored in the variable hist. This will be a one-dimensional NumPy array containing the histogram values for each intensity bin. In this context, each value in the array will represent the number of pixels in the image with a specific intensity value within the specified channel.

```python
#colored image histrograme equlizer
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# Load the color image
img = cv.imread('prop 1.jpg')

# Split the image into color channels
b, g, r = cv.split(img)

# Perform histogram equalization on each channel
equ_b = cv.equalizeHist(b)
equ_g = cv.equalizeHist(g)
equ_r = cv.equalizeHist(r)

# Merge the equalized channels back into a color image
equ = cv.merge((equ_b, equ_g, equ_r))

# Display histograms and equalized images for each channel
channels = [equ_b, equ_g, equ_r]
channel_names = ['Blue Channel', 'Green Channel', 'Red Channel']

for i, channel in enumerate(channels):
    plt.figure(figsize=(10, 6))
    
    hist, bins = np.histogram(channel.ravel(), 256, [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max() / cdf.max()
    
    plt.subplot(2, 2, 1)
    plt.plot(cdf_normalized, color='b')
    plt.hist(channel.flatten(), 256, [0, 256], color='r')
    plt.xlim([0, 256])
    plt.legend(('cdf', 'histogram'), loc='upper left')
    plt.title(f'Histogram of {channel_names[i]} (Equalized)')
    
    plt.subplot(2, 2, 2)
    plt.imshow(channel, cmap='gray')
    plt.axis('off')
    plt.title(f'{channel_names[i]} Image')
    
plt.subplot(2, 2, 3)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Color Image')

plt.subplot(2, 2, 4)
plt.imshow(cv.cvtColor(equ, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Equalized Color Image')

plt.tight_layout()
plt.show()
```

```python
# Gamma and Histograms

%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img_orig = cv.imread('prop 1.jpg', cv.IMREAD_COLOR)

gamma = 3
table = np.array([(i/255.0)**(1.0/gamma)*255.0 for i in np.arange(0,256)]).astype('uint8')
img_gamma = cv.LUT(img_orig, table)

img_orig = cv.cvtColor(img_orig, cv.COLOR_BGR2RGB)
img_gamma = cv.cvtColor(img_gamma, cv.COLOR_BGR2RGB)

f, axarr = plt.subplots(3,2, figsize=(8,8))
axarr[0,0].imshow(img_orig)
axarr[0,1].imshow(img_gamma)

color = ('b', 'g', 'r')
for i, c in enumerate(color):
    hist_orig = cv.calcHist([img_orig], [i], None, [256], [0,256])
    axarr[1,0].plot(hist_orig, color = c)
    hist_gamma = cv.calcHist([img_gamma], [i], None, [256], [0,256])
    axarr[1,1].plot(hist_gamma, color = c)    
    
axarr[2,0].plot(table)
axarr[2,0].set_xlim(0,255)
axarr[2,0].set_ylim(0,255)
axarr[2,0].set_aspect('equal')
```

table = np.array([(i/255.0)**(1.0/gamma)*255.0 for i in np.arange(0,256)]).astype('uint8')
------------------------------------------------
This line creates a lookup table (LUT) that defines how each input pixel value should be transformed based on the chosen gamma value. Here's how this line works:

np.arange(0,256) generates an array containing pixel values from 0 to 255 (256 values in total).

(i/255.0)**(1.0/gamma) calculates the gamma correction transformation for each pixel value. It divides the pixel value by 255 to normalize it to the range [0, 1], applies the inverse gamma transformation, and then scales it back to the [0, 255] range.

The entire expression is multiplied by 255 to scale the output back to the full range of pixel values.

Finally, .astype('uint8') converts the array to an unsigned 8-bit integer data type, which is the standard data type for pixel values in images.

The resulting table is a lookup table that maps each input pixel value to its corresponding gamma-corrected value.

------------------------------------------------------------------------
img_gamma = cv.LUT(img_orig, table)
--------------------------------------------------------------------------
Here, cv.LUT() is used to apply the lookup table to the original input image (img_orig). It performs the pixel-wise transformation according to the LUT, effectively applying the gamma correction to the entire image. The resulting image with gamma correction applied is stored in the img_gamma variable.

In summary, this code snippet calculates a gamma correction lookup table based on the specified gamma value, and then applies this LUT to the original image to achieve the desired gamma correction transformation. This process adjusts the brightness and contrast of the image's mid-tones while leaving the darker and lighter regions relatively unchanged.







```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
gamma = [0.2, 0.5, 1, 2, 5]
for g in gamma:
    table = np.array([(i/255.0)**(g)*255.0 for i in np.arange(0,256)]).astype('uint8')
    ax.plot(table, label=r'$\gamma = {g}$'.format(g=g))
ax.legend(loc='best')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
plt.savefig('transform.png')
plt.show()
```
